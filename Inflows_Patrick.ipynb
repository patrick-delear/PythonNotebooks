# Milky Way with Inflows 

```python
fig = plt.figure(1, figsize=(20,5))
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = fig.add_axes([0.73,0.4,0.2,0.2],frameon=False)

colors = ['Teal','Blue','Gold','Orange']
kwargs = {'tau_in':np.array([1e8,5e8,8e7,3e8]),"tau":np.array([7e8]),'M_gas_init':np.array([7.495e10*0.2])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_inflows_tau_in.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),**kwargs)

ax3.set_xscale('linear')
ax3.set_ylim(bottom=0)
plt.tight_layout()
#plt.savefig('figure1.pdf', format = 'pdf')
plt.show()
```

For an inflows model, I varied tau_in (the inflow rate). By increasing tau_in, we see the biggest variation in the MDF plot. For a longer tau_in, we see a peak of stars being formed at a higher metallicity. As tau_in decreases, we start to see a sharper slope on the sides of the graph, meaning that most of the stars form between two distinct metallicities rather than a gradual increase and decrease of star formation.

```python
fig = plt.figure(1, figsize=(20,5))
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = fig.add_axes([0.73,0.4,0.2,0.2],frameon=False)

colors = ['Teal','Blue','Gold','Red','Green']
kwargs = {'tau_in':np.array([1e8]),'Z_init':np.array([1e-5,7.5e-4,4e-3,9e-7]),'M_gas_init':np.array([7.495e10*0.2]),'tau':np.array([7e8])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_inflows_varyz.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),**kwargs)

colors = ['Gold']
kwargs = {'tau_in':np.array([5e8]),'Z_init':np.array([4e-3]),'M_gas_init':np.array([7.495e10*0.2]),'tau':np.array([7e8])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_inflows_varyz.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),dataon={'elemonfe':False,'MDF':False,'SFR':False,'axlabel':False,'dashedon':True},**kwargs)

ax3.set_xscale('linear')
ax3.set_ylim(bottom=0)
plt.tight_layout()
#plt.savefig('figure1.pdf', format = 'pdf')
plt.show()
```

By changing the initial metallicity (Z_init), we see the knee in the left plot is shifted to the right as Z_init increases. This means that you're going to have a very enriched ISM for a while before enough stars form to increase the amount of Fe, therefore lowering the [Mg/Fe] ratio. The yellow dotted line is the same Z_init as the solid yellow line, but with a tau_in closer to tau. For the MDF plot, I think that the vertical spike is due to the large difference between tau and tau_in. By making the difference smaller, the MDF line becomes more bell-shaped, as seen in the dotted yellow line.

```python
fig = plt.figure(1, figsize=(20,5))
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = fig.add_axes([0.73,0.4,0.2,0.2],frameon=False)

colors = ['Teal','Blue','Gold','Red','Green']
kwargs = {'tau_in':np.array([5e8]),'tau':np.array([7e8]),'M_gas_init':np.array([7.495e10,7.495e10*0.5,7.495e10*0.9,5.27e10,2e9])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_inflows.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),**kwargs)

ax3.set_xscale('linear')
ax3.set_ylim(bottom=0)
plt.tight_layout()
#plt.savefig('figure1.pdf', format = 'pdf')
plt.show()
```

By changing the percentage of your initial (7.495e10) gas mass, you get slight changes in the amount of metal-rich stars towards an earlier time-scale (show in the center plot), meaning that there are an abundance of stars with lower metallicity being formed (compared to the milky way) early. The peak shifting towards the right also means that the number of high metallicity stars is greater than what we see in the milky way.

# Milky Way Inflows and Outflows

```python
fig = plt.figure(1, figsize=(20,5))
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = fig.add_axes([0.73,0.4,0.2,0.2],frameon=False)

colors = ['Teal','Blue','Gold','Red','Green']
kwargs = {'Z_init':np.array([1e-5]),'M_gas_init':np.array([5.27e10]),'eps_out':np.array([6,2,8,4])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_openbox.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),**kwargs)

ax3.set_xscale('linear')
ax3.set_ylim(bottom=0)
plt.tight_layout()
#plt.savefig('figure1.pdf', format = 'pdf')
plt.show()
```

With an open box model (inflows and outflows), and by varying epsilon out, the MDF shifts towards the left or right depending on whether you increase or decrease eps_out respectively. The endpoint stays almost the same though.

```python
fig = plt.figure(1, figsize=(20,5))
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = fig.add_axes([0.73,0.4,0.2,0.2],frameon=False)

colors = ['Teal','Blue','Gold','Red','Green']
kwargs = {'Z_init':np.array([1e-5]),'SNfrac':np.array([0,0.2,0.5,0.7]),'eps_out':np.array([6])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_openbox.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),**kwargs)

ax3.set_xscale('linear')
ax3.set_ylim(bottom=0)
plt.tight_layout()
#plt.savefig('figure1.pdf', format = 'pdf')
plt.show()
```

By increasing the fraction of ejecta that is from SN, the gas in the ISM becomes less enriched over time due to outflows being mainly composed of the enriched gas that would have gone into the ISM. There is also an abundance of metal rich stars being formed early and then then falling off in numbers as [Fe/H] increases (as shown in the center plot)

```python
fig = plt.figure(1, figsize=(20,5))
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = fig.add_axes([0.73,0.4,0.2,0.2],frameon=False)

colors = ['Teal','Blue','Gold','Red','Green']
kwargs = {'Z_init':np.array([4e-3,1e-5,9e-7,7.5e-4]),'tau':np.array([7e8]),'tau_in':np.array([5e8]),'M_gas_init':np.array([5.27e10])}
PLOTmainplots(ax1=ax1,ax2=ax2,ax3=ax3,ax4=ax4,colors=colors,normtopeak=True,fiducial=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','config',"milkyway_openbox.yaml"),directory=os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified'),enginename='sqlite:///'+os.path.join(os.path.expanduser('~'),'Dropbox','Chem_Evol','plots_modified','data','chemevolsimmod.db'),**kwargs)

ax3.set_xscale('linear')
ax3.set_ylim(bottom=0)
plt.tight_layout()
#plt.savefig('figure1.pdf', format = 'pdf')
plt.show()
```

For an open box model and varying initial metallicity, we see similar and smoother plots compared to the inflows only plots. We still see that as the initial metallicity increase, the peak in the MDF shifts to the right and the knee of the left plot extends more to the right as well. This is for the same reasons as before. A higher initial metallicity corresponds to a later abundance of Fe and more stars forming with an [Fe/H] equal to that of the sun.
